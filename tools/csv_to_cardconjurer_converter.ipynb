```python
# Block 1: Define functions

import csv
import json
import os
from IPython.display import display, JSON

def get_frames(cost, show_love):
    """
    Return the appropriate frames based on the card color.
    """
    frame_border = [
        {
            "name": "Black Extension",
            "src": "/img/frames/margins/blackBorderExtension.png",
            "bounds": {
                "x": -0.044,
                "y": -0.02857142857142857,
                "width": 1.088,
                "height": 1.0571428571428572
            },
            "masks": []
        }]

    frame_type, colors = parse_mana_cost(cost)

    frames = create_frame_parts(frame_type, colors, show_love)[::-1] + frame_border
    
    return frames

def parse_mana_cost(cost):
    """
    Parse the mana cost and return a tuple of (frame_type, colors).
    """
    colors = []
    is_hybrid = False
    non_hybrid_count = 0

    for symbol in cost.split('}')[:-1]:
        symbol = symbol.strip('{')
        if symbol.isdigit() or symbol == 'X':
            continue  # Skip generic mana symbols
        elif '/' in symbol:
            is_hybrid = True
            colors.extend(symbol.split('/'))
        else:
            non_hybrid_count += 1
            colors.append(symbol)
    
    unique_colors = list(dict.fromkeys(colors))  # Remove duplicates while preserving order
    
    if not colors:
        return "colorless", []
    elif len(unique_colors) == 1:
        return "mono", unique_colors
    elif len(unique_colors) == 2 and is_hybrid and non_hybrid_count == 0:
        return "hybrid", unique_colors
    else:
        return "multi", unique_colors

def create_frame_parts(frame_type, colors, show_love):
    color_map = {
        'W': 'White',
        'U': 'Blue',
        'B': 'Black',
        'R': 'Red',
        'G': 'Green'
    }
    
    """Create the frame parts based on the frame type and colors."""
    frame_parts = []
    
    if frame_type == 'mono':
        color = colors[0]
        frame_parts.append({
            "name": f"{color_map[color]} Frame",
            "src": f"/img/frames/m15/regular/m15Frame{color}.png",
            "masks": []
        })
    elif frame_type == 'hybrid':
        color1 = colors[0]
        color2 = colors[1]
        frame_parts.extend([
            {
                "name": f"{color_map[color1]} Frame",
                "src": f"/img/frames/m15/regular/m15Frame{color1}.png",
                "masks": []
            },
            {
                "name": f"{color_map[color2]} Frame",
                "src": f"/img/frames/m15/regular/m15Frame{color2}.png",
                "masks": [
                    {
                        "src": "/img/frames/maskRightHalf.png",
                        "name": "Right Half"
                    }
                ]
            }
        ])
    elif frame_type == 'multi':
        frame_parts.append({
            "name": "Gold Frame",
            "src": "/img/frames/m15/regular/m15FrameM.png",
            "masks": []
        })
    # else:  # colorless (this isn't correct)
    #     frame_parts.append({
    #         "name": "Colorless Frame",
    #         "src": "/img/frames/m15/regular/m15FrameColorless.png",
    #         "masks": []
    #     })
    
    if show_love and frame_type == 'mono':
        color = colors[0]
        frame_parts.append({
            "name": f"{color_map[color]} Power/Toughness",
            "src": f"/img/frames/m15/regular/m15PT{color}.png",
            "masks": [],
            "bounds": {
                "height": 0.0733,
                "width": 0.188,
                "x": 0.7573,
                "y": 0.8848
            }
        })
    elif show_love:
        frame_parts.append({
            "name": "Colorless Power/Toughness",
            "src": "/img/frames/m15/regular/m15PTC.png",
            "masks": [],
            "bounds": {
                "height": 0.0733,
                "width": 0.188,
                "x": 0.7573,
                "y": 0.8848
            }
        })
    
    return frame_parts

def create_card_data(card_key, card_code, cost, name, card_type, love, oracle_text, rarity, art, art_x, art_y, art_z, art_r, artist_name, collector_number):
     return {
        "key": card_key,
        "data": {
            "width": 2010,
            "height": 2814,
            "marginX": 0.044,
            "marginY": 0.02857142857142857,
            "frames": get_frames(cost, love != ""),
            "artSource": art,
            "artX": art_x,
            "artY": art_y,
            "artZoom": art_z,
            "artRotate": art_r,
            "setSymbolSource": f"http://localhost:8080/img/setSymbols/custom/a22-{rarity.lower()}.png",
            "setSymbolX": 0.8014925373134328,
            "setSymbolY": 0.5717839374555792,
            "setSymbolZoom": 1.09,
            "watermarkSource": "http://localhost:8080/img/blank.png",
            "watermarkX": 0.33880597014925373,
            "watermarkY": 0.6609808102345416,
            "watermarkZoom": 649,
            "watermarkLeft": "#b79d58",
            "watermarkRight": "none",
            "watermarkOpacity": 0.4,
            "version": "m15Regular",
            "manaSymbols": [],
            "infoYear": "2024",
            "margins": True,
            "bottomInfoTranslate": {
                "x": 0,
                "y": 0
            },
            "bottomInfoRotate": 0,
            "bottomInfoZoom": 1,
            "bottomInfoColor": "white",
            "onload": None,
            "hideBottomInfoBorder": False,
            "showsFlavorBar": True,
            "bottomInfo": {
                "midLeft": {
                    "text": "{elemidinfo-set} • {elemidinfo-language}  {savex}{fontbelerenbsc}{fontsize3}{upinline1}￮{savex2}{elemidinfo-artist}",
                    "x": 0.0647,
                    "y": 0.9548,
                    "width": 0.8707,
                    "height": 0.0171,
                    "oneLine": True,
                    "font": "gothammedium",
                    "size": 0.0171,
                    "color": "white",
                    "outlineWidth": 0.003,
                    "name": "midLeft"
                },
                "topLeft": {
                    "text": "{elemidinfo-rarity} {kerning3}{elemidinfo-number}{kerning0}",
                    "x": 0.0647,
                    "y": 0.9377,
                    "width": 0.8707,
                    "height": 0.0171,
                    "oneLine": True,
                    "font": "gothammedium",
                    "size": 0.0171,
                    "color": "white",
                    "outlineWidth": 0.003,
                    "name": "topLeft"
                },
                "note": {
                    "text": "{loadx}{elemidinfo-note}",
                    "x": 0.0647,
                    "y": 0.9377,
                    "width": 0.8707,
                    "height": 0.0171,
                    "oneLine": True,
                    "font": "gothammedium",
                    "size": 0.0171,
                    "color": "white",
                    "outlineWidth": 0.003,
                    "name": "note"
                },
            },
            "artBounds": {
                "x": 0.0767,
                "y": 0.1129,
                "width": 0.8476,
                "height": 0.4429
            },
            "setSymbolBounds": {
                "x": 0.9213,
                "y": 0.591,
                "width": 0.12,
                "height": 0.041,
                "vertical": "center",
                "horizontal": "right"
            },
            "watermarkBounds": {},
            "text": {
                "mana": {
                    "name": "Mana Cost",
                    "text": cost,
                    "y": 0.0613,
                    "width": 0.9292,
                    "height": 0.03380952380952381,
                    "oneLine": True,
                    "size": 0.043345543345543344,
                    "align": "right",
                    "shadowX": -0.001,
                    "shadowY": 0.0029,
                    "manaCost": True,
                    "manaSpacing": 0
                },
                "title": {
                    "name": "Title",
                    "text": name,
                    "x": 0.0854,
                    "y": 0.0522,
                    "width": 0.8292,
                    "height": 0.0543,
                    "oneLine": True,
                    "font": "belerenb",
                    "size": 0.0381
                },
                "type": {
                    "name": "Type",
                    "text": card_type,
                    "x": 0.0854,
                    "y": 0.5664,
                    "width": 0.8292,
                    "height": 0.0543,
                    "oneLine": True,
                    "font": "belerenb",
                    "size": 0.0324
                },
                "rules": {
                    "name": "Rules Text",
                    "text": oracle_text,
                    "x": 0.086,
                    "y": 0.6303,
                    "width": 0.828,
                    "height": 0.2875,
                    "size": 0.0362
                },
                "pt": {
                    "name": "Power/Toughness",
                    "text": love,
                    "x": 0.7928,
                    "y": 0.902,
                    "width": 0.1367,
                    "height": 0.0372,
                    "size": 0.0372,
                    "font": "belerenbsc",
                    "oneLine": True,
                    "align": "center"
                }
            },
            "infoNumber": collector_number + "/0138",
            "infoRarity": rarity,
            "infoSet": "LUV",
            "infoLanguage": card_code,
            "infoArtist": artist_name,
            "infoNote": "",
            "serialScale": "1"
        }
}

def csv_to_card_conjurer(input_csv, output_file, set_name):
    # Read the CSV file
    with open(input_csv, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        csv_data = list(reader)

    # Create the card conjurer data
    card_conjurer_data = [dummy_card]
    for row in csv_data:
        # Extract all relevant variables from the CSV row
        name = row["Name"]
        if not name:
            continue

        card_key = set_name + " " + row["Card Code"]
        card_code = row["Card Code"]
        cost = row["Cost"]
        card_type = row["Type"]
        oracle_text = row["Oracle Text"]
        flavor_text = row["Flavor Text"]
        love = row["Love"]
        rarity = row["Rarity"][0].upper()

        # Pulls art values, and assigns defaults
        art = f"http://localhost:8080/local_art/{row.get('Art') or 'prototype.png'}"
        art_x = float(row.get('ArtX') or 313) / 2010  # 2010 is card_width per config in create_card_data
        art_y = float(row.get('ArtY') or -593) / 2814  # 2814 card_height per config in create_card_data
        art_z = float(row.get('ArtZ') or 164.2) / 100
        art_r = float(row.get('ArtR') or 11.5)
        artist_name = row.get('Artist') or "Julie Baca"
        

        collector_number = row["Collector #"]

        rules_text = oracle_text + (f"{{flavor}}{flavor_text}" if flavor_text else "")

        card = create_card_data(card_key, card_code, cost, name, card_type, love, rules_text, rarity, art, art_x, art_y, art_z, art_r, artist_name, collector_number)
        card_conjurer_data.append(card)

    card_conjurer_data = card_conjurer_data

    # Debug: Pretty print the data of a specific index
    print(json.dumps(card_conjurer_data[0], indent=2))

    # Write the special format to a file
    with open(output_file, 'w') as f:
        json.dump(card_conjurer_data, f)
```
